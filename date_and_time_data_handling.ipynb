# working with dates


```python
# converting to datetime datatype
date['date']=pd.to_datetime(date['date'])
```

```python
date.info()
```

```python
# 1. Extract year

date['date_year']=date['date'].dt.year
date.sample(5)
```

###2. Extract Month

```python
date['date_month_no']=date['date'].dt.month
```

```python
date.head(2)
```

```python
# extracting month name
date['date_month_name']=date['date'].dt.month_name()
```

```python
# day of week
date['date_dow']=date['date'].dt.dayofweek
date.head(2)
```

```python
# day of week - name
date['date_dow_name']=date['date'].dt.day_name()
date.head(2)
```

```python
# is weekend
date['data_is_weekend']=np.where(date['date_dow_name'].isin(['Sunday','Saturday']),1,0)
```

```python
date.head()
```

```python
# extract week
date['date_week']=date['date'].dt.isocalendar().week
```

```python
date.head()
```

```python
date.drop(columns=(['product_id','city_id','orders','data_is_weekend']))
```

```python
#extract quarter
date['date_quater']=date['date'].dt.quarter
```

```python
#extrat semester
date['semester']=np.where(date['date_quater'].isin([1,2]),1,2)
```
